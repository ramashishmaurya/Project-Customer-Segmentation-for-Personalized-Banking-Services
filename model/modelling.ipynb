```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
df = pd.read_csv('merged.csv')
```

```python
df.head()
```

```python
df['txn_date'] = pd.to_datetime(df['txn_date'])
```

```python
df.head()
```

```python
df['txn_date'] = pd.to_datetime(df['txn_date']).dt.date
```

```python
txn_summary = df.groupby('customer_id').agg({
    'txn_amount': ['sum', 'mean', 'count'],
    'txn_date': lambda x: (pd.Timestamp.now() - pd.to_datetime(x.max())).days
})
```

```python
txn_summary.head()
```

```python
txn_summary.columns
```

```python
txn_summary.columns = ['total_spent', 'avg_spent', 'txn_count', 'recency']
txn_summary.reset_index(inplace=True)
```

```python
txn_summary
```

```python
txn_summary.reset_index(inplace=True)
```

```python
txn_summary
```

```python
customer_info = df.drop_duplicates('customer_id')[['customer_id', 'age', 'income']]
```

```python
customer_info
```

```python
merged = pd.merge(customer_info, txn_summary, on='customer_id')
```

```python
merged.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
import pickle
```

```python
X = merged[['age', 'income', 'total_spent', 'avg_spent', 'txn_count', 'recency']]
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
merged['segment'] = kmeans.fit_predict(X_scaled)
```

```python
merged.head()
```

```python
with open("kmeans_model.pkl", "wb") as f:
    pickle.dump(kmeans, f)
```

```python
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

```python
merged.to_csv("customers_with_segments.csv", index=False)
```
